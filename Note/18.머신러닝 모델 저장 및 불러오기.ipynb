### iris 데이터 불러오기

```python
import pandas as pd
```

```python
iris = pd.read_csv("../Data/iris.csv")

iris.head()
```

### Feature와 Target 분리

```python
iris_data = iris.iloc[:, 0:4] 
iris_target = iris['Name']

print(iris_data.head())
print('--------------------------------')
print(iris_target.head())
print('--------------------------------')
print(iris_data.shape)
print('--------------------------------')
print(iris_target.shape)
```

### train과 test 분리

```python
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target = train_test_split(
    iris_data,
    iris_target,
    random_state=42,
    stratify=iris_target
)

print(train_input.shape)
print('--------------------------------')
print(test_input.shape)
print('--------------------------------')
print(train_target.shape)
print('--------------------------------')
print(test_target.shape)
```

### SVM

```python
from sklearn.svm import SVC
```

```python
svm_clf =SVC()
```

```python
svm_clf.fit(train_input, train_target)
print(svm_clf.score(train_input, train_target))
print(svm_clf.score(test_input, test_target))
```

```python
# 전체 Data로 모델 만들기
svm_clf.fit(iris_data, iris_target)
```

### AI 저장하기

```python
import joblib
```

```python
fileName = '../Model/svm_iris.h5'
joblib.dump(svm_clf, fileName)
```

---

### RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_clf = RandomForestClassifier()
```

```python
rf_clf.fit(train_input, train_target)
print(rf_clf.score(train_input, train_target))
print(rf_clf.score(test_input, test_target))
```

```python
rf_clf.fit(iris_data, iris_target)
```

```python
fileName = '../Model/rf_iris.h5'
joblib.dump(rf_clf, fileName)
```

---

### kNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_clf = KNeighborsClassifier()
```

```python
knn_clf.fit(train_input, train_target)
print(knn_clf.score(train_input, train_target))
print(knn_clf.score(test_input, test_target))
```

```python
knn_clf.fit(iris_data, iris_target)
```

```python
fileName = '../Model/knn_iris.h5'
joblib.dump(knn_clf, fileName)

```

### 저장된 모델 불러와 예측
- SepalLength: 5.1
- SepalWidth: 3.0
- PetalLength: 1.3
- PetalWidth: 0.2

```python
# 데이터 만들기
pred_data = {
    'SepalLength': [5.1],
    'SepalWidth': [3.0],    
    'PetalLength': [1.3],
    'PetalWidth': [0.2]
}

df = pd.DataFrame(pred_data)
df
```

```python
# svm
svm_clf = joblib.load('../Model/svm_iris.h5')
print(svm_clf.predict(df))

# rf
rf_clf = joblib.load('../Model/rf_iris.h5')
print(rf_clf.predict(df))

# knn
knn_clf = joblib.load('../Model/knn_iris.h5')
print(knn_clf.predict(df))
```
